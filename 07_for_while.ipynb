```python
for a in range(1, 10):
    for i in range(1, 10):
        print(f"{a}*{i}={a*i}  ", end='')
    print()
```

```python
a = 1
while a<10:
    for i in range(1, 10):
        print(f"{a}*{i}={a*i}  ", end='')
    print()
    a += 1
```

```python
su = int(input('몇 단을 출력해볼까요? 예) 9단 -> 9   '))
i = 1
while i <= su:
    if not i % 11:
        print()
    print(f"{su}*{i}={su*i}  ", end='')
    i += 1
print('\n' + "#"*40)
```
